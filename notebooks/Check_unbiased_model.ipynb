```python
n_rows = 2
n_bins = 3
n_cols = len(ss2s) // n_rows
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 4*n_rows))

for i, c in enumerate(ss2s):
    row = i // n_cols
    col = i % n_cols
    ax = axes[row, col] if n_rows > 1 else axes[col]

    # Calculate errors
    errors = df[c] - df[c + '_predictions']
    
    # Create bins for grouping ground truth values
    bins = np.linspace(df[c].min(), df[c].max(), n_bins + 1)
    ground_truth_range = pd.cut(df[c], bins=bins)

    # Compute the variance of errors for each ground truth range
    variance_by_range = errors.groupby(ground_truth_range).std()

    # Plot the variance of errors
    variance_by_range.plot(kind='bar', color='skyblue', ax=ax)
    ax.set_title(f'{ss2s[c]}')
    ax.set_xlabel('Ground Truth Range')
    ax.set_ylabel('Variance of Errors')
    ax.grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
n_cols = len(ss2s) // n_rows
fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 4*n_rows))

for i, c in enumerate(ss2s):
    row = i // n_cols
    col = i % n_cols
    ax = axes[row, col] if n_rows > 1 else axes[col]

    # Calculate errors
    errors = df[c + '_predictions'] - df[c]
    
    # Create bins for grouping ground truth values
    bins = np.linspace(df[c].min(), df[c].max(), n_bins + 1)
    ground_truth_range = pd.cut(df[c], bins=bins)

    # Compute the mean and standard deviation of errors for each ground truth range
    mean_by_range = errors.groupby(ground_truth_range).mean()
    std_by_range = errors.groupby(ground_truth_range).std()
    count_by_range = errors.groupby(ground_truth_range).count()

    # Plot the mean of errors with error bars and density of each bin
    ax.bar(mean_by_range.index.astype(str), mean_by_range.values, yerr=std_by_range.values, color='skyblue', capsize=5)
    ax.set_title(f'{ss2s[c]}')
    ax.set_xlabel('Ground Truth Range')
    ax.set_ylabel('Prediction Errors')
    ax.grid(axis='y')

    # Add text for the density of each bin
    for bar, count in zip(ax.patches, count_by_range):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width() * 1.5 / 2, height + 0.01, f'{count}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
